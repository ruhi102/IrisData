```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
from sklearn import datasets
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
# import the Iris data set

iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
# set up some stuff for the plots

Xzeros0 = (0. * X) + .2
Xzeros1 = (0. * X) 
Xzeros2 = (0. * X) - .2

x_min, x_max = X.min() - 0.5, X.max() + 1.5
```

```python
# show the data

for i in range (len(Y)):
    print (i, X[i,:],Y[i])
```

```python
# get Iris classifications

IC = np.unique(Y)
IrisC = ('Setosa','Versicolor','Virginica')
pltC = ('red', 'blue', 'green')
pltCback = ('pink', 'lightblue','lightgreen')
pltM = ('o', 'x', '*')
for i in range(0,3):
    print (IC[i],pltM[i],IrisC[i],pltC[i],pltCback[i])
          
```

```python
# setup Iris feature lablels

IrisF=("sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)")
for i in range(0,4):
    print(IrisF[i])
```

```python
# plot data for all pairs of features 

plt.close('all')
fig, ax = plt.subplots(nrows=3, ncols=2, figsize= (8,12), sharex=True, sharey=True)

i = 0
k = 0
k1 = 0
k2 = 0
while i < 4:
    x2F = IrisF[i]
    j = 0
    while j < i:
        k1 = np.int(k/2)
        k2 = np.mod(k,2)
        x1F = IrisF[j]
        ax[k1,k2].scatter(X[Y == IC[0], j], X[Y == IC[0], i], 
                          color=pltC[0], marker=pltM[0], label=IrisC[0])
        ax[k1,k2].scatter(X[Y == IC[1], j], X[Y == IC[1], i], 
                          color=pltC[1], marker=pltM[1], label=IrisC[1])
        ax[k1,k2].scatter(X[Y == IC[2], j], X[Y == IC[2], i], 
                          color=pltC[2], marker=pltM[2], label=IrisC[2])
        ax[k1,k2].set(aspect='equal')
        ax[k1,k2].set_xlim(x_min,x_max)
        ax[k1,k2].set_ylim(x_min,x_max)
        ax[k1,k2].set_xlabel(x1F)
        ax[k1,k2].set_ylabel(x2F)
        ax[k1,k2].legend(loc='upper right', fontsize='x-small')
        j += 1
        k += 1
    i += 1
plt.show() 
```

```python
cdx = np.array([0,1,2])   # this array contains the classes to be considered (can be 2-3 classes)
 
from itertools import combinations
fdx = np.array([0,1,2,3])           # this array contains the features to be considered (can be 1-4 features)
r = 2                               # this is used to get the 2 different features to be implemented in the different combinations generated 
list_com = list(combinations(fdx, r)) #generate different combinations of the list similar different number sequence 

print (list_com)
nC = len(cdx)
nF = len(fdx)

print ('\nClasses')
for i in range(nC):
    print (cdx[i],IrisC[cdx[i]])
    
print ('\nFeatures')
for i in range(nF):
    print (fdx[i],IrisF[fdx[i]])
```

```python
# build the XX and YY arrays
# these arrays contain the samples in the classes specified in cdx and include the features specified in fdx

i = 0
numY = 0
for i in range(len(Y)):
    for j in range(nC):
        if Y[i] == IC[cdx[j]]:
            numY += 1

YY = np.zeros((numY))
Array = []
#loop through the combinations and import the data here to be able to find the accuracy scores
for fdx in list_com:
    nF = len(fdx)
    XX = np.zeros((numY,nF))
    if (len(fdx) == 1):
        XXzeros0 = (0. * XX) + .2
        XXzeros1 = (0. * XX) 
        XXzeros2 = (0. * XX) - .2

    j = 0
    for i in range(len(Y)):
        for k in range(nC):
            if Y[i] == IC[cdx[k]]:        
                YY[j] = IC[cdx[k]]
                for m in range(nF):
                    XX[j,m] = X[i,fdx[m]] 
                j += 1
    Array.append(XX)  #append the data found from the combinations into an array to be able to generate the scores
print(Array)
```

```python
# Feature 1: (0, 1), Feature 2: (0, 2), Feature 3: (0, 3), Feature 4: (1, 2), Feature 5: (1, 3), Feature 6: (2, 3)

#found online by searching Scikit learn on how to implement the loss function parameter here.
clf = SGDClassifier(tol=1e-4, random_state=0, loss = 'modified_huber')

count = 0
print('\n Modified Huber Loss:')
for XX in Array:
    clf.fit(XX,YY)
    count = count + 1
    converted_num = str(count)
    print('\n Feature ' + converted_num + ' Score: ',clf.score(XX,YY))
```

```python
# Feature 1: (0, 1), Feature 2: (0, 2), Feature 3: (0, 3), Feature 4: (1, 2), Feature 5: (1, 3), Feature 6: (2, 3)

#found online by searching Scikit learn on how to implement logistic regression
clf1 = LogisticRegression(tol=1e-4, random_state=0)
count = 0
print('\n Logistic Regresion:')
for XX in Array:
    clf1.fit(XX,YY)
    count = count + 1
    converted_num = str(count)
    print('\n Feature ' + converted_num + ' Score: ',clf1.score(XX,YY))
```

```python
# Feature 1: (0, 1), Feature 2: (0, 2), Feature 3: (0, 3), Feature 4: (1, 2), Feature 5: (1, 3), Feature 6: (2, 3)

#found online by searching Scikit learn on how to implement support vector machine and this time using parameter kernal to 
# implement linear classification
linear1 = SVC(kernel = 'linear')
count = 0
print('\n Support Vector Machine(Linear):')
for XX in Array:
    linear1.fit(XX,YY)
    count = count + 1
    converted_num = str(count)
    print('\n Feature ' + converted_num + ' Score: ',linear1.score(XX,YY))
```

```python
# Feature 1: (0, 1), Feature 2: (0, 2), Feature 3: (0, 3), Feature 4: (1, 2), Feature 5: (1, 3), Feature 6: (2, 3)

#found online by searching Scikit learn on how to implement support vector machine and this time using parameter kernal to 
# implement polynomial classification
poly1 = SVC(kernel = 'poly')
count = 0
print('\n Support Vector Machine(Polynomial):')
for XX in Array:
    poly1.fit(XX,YY)
    count = count + 1
    converted_num = str(count)
    print('\n Feature ' + converted_num + ' Score: ',poly1.score(XX,YY))
```

```python
# Feature 1: (0, 1), Feature 2: (0, 2), Feature 3: (0, 3), Feature 4: (1, 2), Feature 5: (1, 3), Feature 6: (2, 3)

#found online by searching Scikit learn on how to implement support vector machine and this time using parameter kernal to 
# implement rbf classification
rbf1 = SVC(kernel = 'rbf')
count = 0
print('\n Support Vector Machine(RBF):')
for XX in Array:
    rbf1.fit(XX,YY)
    count = count + 1
    converted_num = str(count)
    print('\n Feature ' + converted_num + ' Score: ',rbf1.score(XX,YY))
```
