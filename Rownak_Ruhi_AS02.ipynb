```python
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
from sklearn import datasets
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
```

```python
iris = datasets.load_iris()
X = iris.data
Y = iris.target
```

```python
Xzeros0 = (0. * X) + .2
Xzeros1 = (0. * X) 
Xzeros2 = (0. * X) - .2

x_min, x_max = X.min() - 0.5, X.max() + 1.5
```

```python
# show the data

for i in range (len(Y)):
    print (i, X[i,:],Y[i])
```

```python
IC = np.unique(Y)
IrisC = ('Setosa','Versicolor','Virginica')
# pltC = ('red', 'blue', 'green')
# pltCback = ('pink', 'lightblue','lightgreen')
# pltM = ('o', 'x', '*')
for i in range(0,3):
    print (IC[i],IrisC[i])
```

```python
IrisF=("sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)")
for i in range(0,4):
    print(IrisF[i])
```

```python
#find a way where you can get all 4 combinations of classes in array format

cdx = np.array([0,1,2])             # this array contains the classes to be considered (can be 2-3 classes)
#imported combinations from itertools found from online
from itertools import combinations
fdx = np.array([0,1,2,3], dtype=object)  # this array contains the features to be considered (can be 1-4 features)
list_com = list()     #this is where the list would be stored
for n in range(len(fdx) + 1):      #loop through the length of fdx which has all the features and add one more
    list_com += list(combinations(fdx, n)) #generate different combinations of the list similar different number sequence 
                                           #and each list would be different combinations of features
print(list_com) # print out the list 


print(cdx)
# print(fdx)

nC = len(cdx)
nF = len(fdx)


print ('\nClasses')
for i in range(nC):
    print (cdx[i],IrisC[cdx[i]])
    
print ('\nFeatures')
for i in range(nF):
    print (fdx[i],IrisF[fdx[i]])
```

```python
i = 0
numY = 0
for i in range(len(Y)):
    for j in range(nC):
        if Y[i] == IC[cdx[j]]:
            numY += 1

YY = np.zeros((numY))
Array = []
#loop through the combinations and import the data here to be able to find the accuracy scores
#start from 1 otherwise it would return an empty array in the beginning
for fdx in list_com[1:]:
    nF = len(fdx)
    XF = np.zeros((numY,nF))
    if (len(fdx) == 1):
        XFzeros0 = (0. * XF) + .2
        XFzeros1 = (0. * XF) 
        XFzeros2 = (0. * XF) - .2

    j = 0
    for i in range(len(Y)):
        for k in range(nC):
            if Y[i] == IC[cdx[k]]:        
                YY[j] = IC[cdx[k]]
                for m in range(nF):
                    XF[j,m] = X[i,fdx[m]] 
                j += 1
    Array.append(XF)
print(Array)
            
```

```python
# fit the model

# clf = Perceptron(tol=1e-4, random_state=0)

#found online by searching Scikit learn on how to implement the loss function parameter here.
clf = SGDClassifier(tol=1e-4, random_state=0, loss = 'hinge')

for XF in Array:
    clf.fit(XF,YY)

    print('\nScore: ',clf.score(XF,YY))
```

```python
#implemented the second loss function which is perceptron
clf1 = SGDClassifier(tol=1e-4, random_state=0, loss = 'perceptron')

for XF in Array:
    clf1.fit(XF,YY)

    print('\nScore: ',clf1.score(XF,YY))
```

```python
#implemented the third loss function which is modified huber 
clf2 = SGDClassifier(tol=1e-4, random_state=0, loss = 'modified_huber')

for XF in Array:
    clf2.fit(XF,YY)

    print('\nScore: ',clf2.score(XF,YY))
```

```python
#implemented the last loss function which is log
clf3 = SGDClassifier(tol=1e-4, random_state=0, loss = 'log')

for XF in Array:
    clf3.fit(XF,YY)

    print('\nScore: ',clf3.score(XF,YY))
```
